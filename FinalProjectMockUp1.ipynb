```python
# Importing dependencies
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn import metrics

import warnings
warnings.filterwarnings('ignore')
```

```python
# Exploratory analysis on META data using column 'Volume'
plt.figure(figsize=(20,5))
plt.plot(meta_df['Volume'])
plt.title('Meta Stock Volume')
plt.ylabel('No. of shares')
plt.show()
```

```python
# Exploratory analysis on META data using column 'Close'
plt.figure(figsize=(20,5))
plt.plot(meta_df['Close'])
plt.title('Meta Closing Price')
plt.ylabel('Price in dollars')
plt.show()
```

```python
# Checking dataframe info.
meta_df.info()
```

```python
# Checking for null values.
meta_df.isnull().sum()
```

```python
# Adding columns for training purposes and adding a target variable.
meta_df['open-close'] = meta_df['Open'] - meta_df['Close']
meta_df['low-high'] = meta_df['Low'] - meta_df['High']
meta_df['target'] = np.where(meta_df['Close'].shift(-1) > meta_df['Close'],1,0)
```

```python
# Check if target variable is balanced.
plt.pie(meta_df['target'].value_counts().values,labels=[0,1], autopct='%1.1f%%')
plt.show()
```

```python
# Data splitting and normalization
features = meta_df[['open-close','low-high']]
target = meta_df['target']

scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_test, y_train, y_test = train_test_split(
    features, target, random_state=1)

print(X_train.shape, X_test.shape)
```

```python
# Training the models - LogisticRegression, SVC, XGBClassifier
models = [LogisticRegression(), SVC(kernel='poly', probability=True), XGBClassifier()]

for i in range(3):
  models[i].fit(X_train, y_train)

  print(f'{models[i]} : ')
  print('Training Accuracy: ',metrics.roc_auc_score( y_train, models[i].predict_proba(X_train)[:,1]))
  print(f'{models[i]} : ')
  print('Validation Accuracy: ',metrics.roc_auc_score( y_test, models[i].predict_proba(X_test)[:,1]))
  print()
```

```python
# Plotting the confusion matrix.
metrics.plot_confusion_matrix(models[0], X_test, y_test)
plt.show()
```
